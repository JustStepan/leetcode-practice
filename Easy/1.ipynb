# Условие задачи.
Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

 

Example 1:

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
Example 2:

Input: nums = [3,2,4], target = 6
Output: [1,2]
Example 3:

Input: nums = [3,3], target = 6
Output: [0,1]

## Подход с pairwise
Этот инструмент itertools позволяет безопасно иетрироваться по соседним элементам итерируемого объекта (нет ошибки выхода за пределы списка). 
Здесь я обернул его в enumerate для того, чтобы получать не просто значения (их сумму), но и индексы. 
### Проблемы подхода. 
когда элементы стоят не рядом - не сработает. 
print("fourth:", twoSum([3,2,3], 6))

```python
from itertools import pairwise

def twoSum(nums, target):
    for index, (current, next_item) in enumerate(pairwise(nums)):
        if current + next_item == target:
            return [index, index + 1]

print("first:", twoSum([2,7,11,15], 9))
print("second:", twoSum([3,2,4], 6))
print("third:", twoSum([3, 3], 6))
print("fourth:", twoSum([3,2,3], 6))
```

## Подход ВЛОБ. 
ЛОгика такая: итерируемся дважды по одному и тому же списку. Если одинаковые индексы пропускаем. если target найден возвращаем индексы
### Проблема.
ПРоходим несколько раз по одним и тем же значениям.  

```python
%%time
def twoSum(nums, target):
    for index1, num in enumerate(nums):
        for index2, secnum in enumerate(nums):
            if index1 == index2:
                continue
            if num + secnum == target:
                return [index1, index2]

print("first:", twoSum([2,7,11,15], 9))
print("second:", twoSum([3,2,4], 6))
print("third:", twoSum([3, 3], 6))
print("fourth:", twoSum([3,2,3], 6))
```

## Попытка исключить уже проитерированные числа. 
для этого второй раз итерируемся по срезу.
 этом случае код 

            if index1 == index2:
                continue

не актуальный

```python
%%time
def twoSum(nums, target):
    for index1, num in enumerate(nums):
        for index2, secnum in enumerate(nums[index1 + 1:]):
            if num + secnum == target:
                return [index1, index2 + index1 + 1]

print("first:", twoSum([2,7,11,15], 9))
print("second:", twoSum([3,2,4], 6))
print("third:", twoSum([3, 3], 6))
print("fourth:", twoSum([3,2,3], 6))
print("fifth:", twoSum([3,2,3, 4, 11], 13))
```

## Лучшее решение - делать через запись в словарь значений.
логика: делаем словарь {значение: индекс}, итерируемся с помощью enumerate, если разница target и значения в словаре найдена возвращаем индекс из словаря и индекс enumerate если не найдена записываем в словарь. 

```python
# %%time
def twoSum(nums, target):
    value_index_dict = {}

    for index, num in enumerate(nums):
        diff = target - num
        if diff in value_index_dict:
            return [value_index_dict[diff], index]

        value_index_dict[num] = index


print("first:", twoSum([2,7,11,15], 9))
print("second:", twoSum([3,2,4], 6))
print("third:", twoSum([3, 3], 6))
print("fourth:", twoSum([3,2,3], 6))
print("fifth:", twoSum([3,2,3, 4, 11], 13))
```
