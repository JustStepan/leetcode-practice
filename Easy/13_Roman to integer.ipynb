# Roman to integer
## Description
Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.

Symbol       Value
I             1
V             5
X             10
L             50
C             100
D             500
M             1000
For example, 2 is written as II in Roman numeral, just two ones added together. 12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II.

Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not IIII. Instead, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances where subtraction is used:

I can be placed before V (5) and X (10) to make 4 and 9. 
X can be placed before L (50) and C (100) to make 40 and 90. 
C can be placed before D (500) and M (1000) to make 400 and 900.
Given a roman numeral, convert it to an integer.

```python
from itertools import pairwise

def romanToInt(s: str):

    rim_nums = {
        "I": 1,
        "V": 5,
        "X": 10,
        "L": 50,
        "C": 100,
        "D": 500,
        "M": 1000,
    }

    stops = ("IV", "IX", "XL", "XC", "CD", "CM")
    summ = 0
    for before, after in pairwise(s):
        if (before + after) in stops:
            summ -= rim_nums[before]
        else:
            summ += rim_nums[before]
    summ += rim_nums[s[-1]]
    return summ
        
print(romanToInt('XIX'))
print(romanToInt('MCMXCIV'))
print(romanToInt('MCDLXXVI'))

```

Это самое распространенное решение. На leetcode время выполнения разнится от 0мс до 15 мс.  Оценка ИИ - хорошее и современное решени! Впрочем ИИшка, кнечно, предложил вариант "лучше": см. ниже. ТЕм не менее этот вариант на платформе ничем практически не отличается по времени выполнения. 

```python
def romanToInt(s: str):
    rim_nums = {
        "I": 1, "V": 5, "X": 10, "L": 50,
        "C": 100, "D": 500, "M": 1000,
    }

    total = 0
    # Проходим по индексам до предпоследнего элемента
    for i in range(len(s) - 1):
        current_val = rim_nums[s[i]]
        next_val = rim_nums[s[i+1]]
        
        if current_val < next_val:
            total -= current_val
        else:
            total += current_val
            
    # Добавляем последний символ
    total += rim_nums[s[-1]]
    return total

print(romanToInt('XIX'))
print(romanToInt('MCMXCIV'))
print(romanToInt('MCDLXXVI'))
```
