# 9. Palindrome number
Given an integer x, return true if x is a palindrome, and false otherwise.
Example 1:

Input: x = 121
Output: true
Explanation: 121 reads as 121 from left to right and from right to left.
Example 2:

Input: x = -121
Output: false
Explanation: From left to right, it reads -121. From right to left, it becomes 121-. Therefore it is not a palindrome.
Example 3:

Input: x = 10
Output: false
Explanation: Reads 01 from right to left. Therefore it is not a palindrome.

# Решение ВЛОБ
очевидно, что можно сравнивать списки из строк и получать результат. Но это не лучшее решение согласно leetcode: примерно 50%. 

```python
%time
def isPalindrome(x: int) -> bool:
    if list(str(x)) == list(str(x))[::-1]:
        return True
    else:
        return False

print(isPalindrome(121))
print(isPalindrome(-121))
print(isPalindrome(10))
```

```python
%time
def isPalindrome(x: int) -> bool:
    if x < 0:
        return False

    x_list = list(str(x))

    if len(str(x)) % 2 == 0:
        l = len(str(x))/2
    else:
        l = (len(str(x)) - 1)/2
        
    count = 0
    for i in range(int(l)):
        if x_list[count] != x_list[-count - 1]:
            return False
        count += 1
    return True

print(isPalindrome(121))
print(isPalindrome(-121))
print(isPalindrome(10))
```

# Сравнения конца и начала
Прошлый раз сравнение было строк, а теперь будем через "//" и "%" сравнивать числа (int). Это должно быть лучшее решение.
## UPD 
думал это решение будет лучшим но нет это решение оказалось даже хуже первого. ПОчему? Вот: **Скорость** (O(N)): Квадратные корни 10 ** (lenght - 1) внутри цикла — это очень дорогостоящая операция. Это делает алгоритм значительно медленнее строковых методов.

```python
%time
def isPalindrome(x: int) -> bool:
    # first get away from negative numbers
    if x < 0:
        return False
    lenght = len(str(x))
    while lenght > 1: 
        if x // (10 ** (lenght - 1)) != x % 10:
            return False
        x = (x % (10 ** (lenght - 1))) // 10
        lenght -= 2
    return True
    

print(isPalindrome(121))
print(isPalindrome(-121))
print(isPalindrome(10))
print(isPalindrome(0))
print(isPalindrome(123456789))
```

# Лучший Вариант 
Это улучшенный второй вариант (по читаемости и логике). Хотя leetcode поазывает не самый выдающийся результат


```python
def isPalindrome(x: int) -> bool:
    if x < 0:
        return False
    # Строка в Python позволяет получить доступ по индексу s[i]
    s = str(x)
    left, right = 0, len(s) - 1
    
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
        
    return True

print(isPalindrome(121))
print(isPalindrome(-121))
print(isPalindrome(10))
print(isPalindrome(0))
print(isPalindrome(123456789))
```
