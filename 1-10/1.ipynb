# Условие задачи.
Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

 

Example 1:

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
Example 2:

Input: nums = [3,2,4], target = 6
Output: [1,2]
Example 3:

Input: nums = [3,3], target = 6
Output: [0,1]

## Подход с pairwise
Этот инструмент itertools позволяет безопасно иетрироваться по соседним элементам итерируемого объекта (нет ошибки выхода за пределы списка). 
Здесь я обернул его в enumerate для того, чтобы получать не просто значения (их сумму), но и индексы. 
### Проблемы подхода. 
когда элементы стоят не рядом - не сработает. 
print("fourth:", twoSum([3,2,3], 6))

```python
from itertools import pairwise

def twoSum(nums, target):
    for index, (current, next_item) in enumerate(pairwise(nums)):
        if current + next_item == target:
            return [index, index + 1]

print("first:", twoSum([2,7,11,15], 9))
print("second:", twoSum([3,2,4], 6))
print("third:", twoSum([3, 3], 6))
print("fourth:", twoSum([3,2,3], 6))
```

## Подход с поиском индекса. 
ЛОгика такая: ищем разность между target и текущим значением (хотя можно искать и сумму - не суть), итерируемся и ищем индекс. Если совпадение есть - возвращаем индекс текущий и найденный. Для этого нужно итерироваться по всему списку. 

```python
def twoSum(nums, target):
    for index, num in enumerate(nums):
        for secnum in nums[index:]:
            print(nums[:index])
            if num + secnum == target:
                return [index, nums.index(secnum)]

print("first:", twoSum([2,7,11,15], 9))
print("second:", twoSum([3,2,4], 6))
print("third:", twoSum([3, 3], 6))
print("fourth:", twoSum([3,2,3], 6))
```
